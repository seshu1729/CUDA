```python
%%cuda
#include <stdio.h>

__global__ void helloCUDA() {
    // blockIdx.x: Which block is this?
    // threadIdx.x: Which thread inside the block is this?
    // blockDim.x: How many threads are in one block?

    int globalID = blockIdx.x * blockDim.x + threadIdx.x;

    printf("Hello from Global ID %d [Block %d, Thread %d]\n",
           globalID, blockIdx.x, threadIdx.x);
}

int main() {
    // Launch configuration: 2 blocks, 4 threads each
    // Total of 8 threads will run
    printf("Launching GPU kernels...\n");

    helloCUDA<<<2, 4>>>();

    // Crucial: Synchronize so the CPU waits for the GPU printf to finish
    cudaDeviceSynchronize();

    printf("GPU execution finished.\n");
    return 0;
}
```
